```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Number of players
num_players = 20
player_names = [
    "Cristiano Ronaldo", "Lionel Messi", "Kylian Mbappé", "Erling Haaland",
    "Kevin De Bruyne", "Robert Lewandowski", "Neymar Jr.", "Mohamed Salah",
    "Virgil van Dijk", "Harry Kane", "Raheem Sterling", "N'Golo Kanté",
    "Joshua Kimmich", "Alisson Becker", "Ederson Moraes", "Jan Oblak",
    "Trent Alexander-Arnold", "Phil Foden", "Luka Modrić", "Karim Benzema"
]
# Create a DataFrame with random data
df = pd.DataFrame({
    'Player Name':player_names,
    'Number of Goals': np.random.randint(0, 30, size=num_players),
    'Number of Assists': np.random.randint(0, 20, size=num_players),
    'Yellow Cards': np.random.randint(0, 5, size=num_players),
    'Red Cards': np.random.randint(0, 2, size=num_players)
})
mean_goals = df['Number of Goals'].mean()
df['Mean Goals'] = mean_goals
Std_goals = df['Number of Goals'].std()
df['Standard deviation Goals'] = Std_goals
plt.figure(figsize=(15, 6))
plt.bar(df['Player Name'], df['Number of Goals'])
plt.xlabel('Player Name')
plt.ylabel('Number of Goals')
plt.title('Number of Goals per Player')
plt.xticks(rotation=90)
plt.show()

```
