```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data=[12,15,14,10,18,20,22,17,19,13,16,15,14,21,23]
mean_d=np.mean(data)
print("mean",mean_d)
median_d=np.median(data)
print("median",median_d)
std_d=np.std(data)
print("standard deviation",std_d)
variance_d=np.var(data)
print("variance",variance_d)
range_d=np.max(data)-np.min(data)
print("range",range_d)
minus_std=mean_d-std_d
plus_std=mean_d+std_d
print("mean -standard devaition=",minus_std)
print("mean -standard devaition=",plus_std)
plt.hist(bins=10,x=data)
plt.xlabel('data')
plt.ylabel('frequency')
plt.title('histogram')
plt.axvline(x=mean_d,color='red',linestyle='--',label='mean')
plt.legend()
plt.axvline(x=minus_std,color='green',linestyle='--',label='minus_std')
plt.legend()
plt.axvline(x=plus_std,color='green',linestyle='--',label='plus_std')
plt.legend()
plt.axvspan(minus_std, plus_std, alpha=0.2, color='yellow', label='±std')
plt.legend()
plt.show()
```

The histogram has a roughly symmetric distribution, meaning that the data is evenly spread around the center.
there are outliers outside the one standard deviation range shown in yellow.
