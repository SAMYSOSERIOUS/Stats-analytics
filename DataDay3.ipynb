load Iris

```python
import seaborn as sns
iris = sns.load_dataset('iris')
print(iris.head())
```

sample Iris (30 observations)

```python
sample = iris.sample(n=30)
sample
```

sample mean ditribution Analysis 100 times for the mean sepal length and plot as a histogram

```python
import matplotlib.pyplot as plt

def radomize_n(n):
  means=[]
  for i in range(n):
    sample = iris.sample(n=30)
    mean=sample.sepal_length.mean()
    means.append(mean)
  return means

print("write number of samplings")
n=int(input())
meanslist=radomize_n(n)
print(meanslist)
plt.hist(meanslist)
plt.title("mean distribution")
plt.xlabel("mean")
plt.ylabel("frequency")
plt.show()
```

systemic sampling by taking 20 percent of dataset then plot original dataset and sampled data

```python
import matplotlib.pyplot as plt
import numpy as np
print("systemic sampling")
def systematic_sampling(df,step):
  indexes=np.arange(0,len(df),step=step)
  systemic_sample=df.iloc[indexes]
  return systemic_sample
print(len(iris))
print("write percentage :")
percentage=float(input())/100
systemic_sample=systematic_sampling(iris,1/percentage)
print(systemic_sample)
plt.hist(x=systemic_sample.sepal_length,color="red",alpha=0.7)
plt.hist(x=iris.sepal_length,color="blue",alpha=0.3)
plt.title("sepal length")
plt.xlabel("sepal length")
plt.ylabel("frequency")
plt.show()
```
